# Rotating Hexagon Simulation
This notebook creates a simulation of a ball inside a rotating hexagon with realistic physics and writes the output to an HTML file.

## Setup Output Directory
Ensure the 'output' folder exists and is ready to store the generated HTML file.

```python
# Setup Output Directory
import os

output_dir = "output"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
print(f"Output directory '{output_dir}' is ready.")
```

## Generate HTML Structure
Create the basic HTML structure, including a canvas element for rendering the animation.

```python
# Generate HTML Structure
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Rotating Hexagon Simulation</title>
    <link rel="stylesheet" href="styles.css">
</head>
<body>
    <canvas id="simulationCanvas"></canvas>
    <script src="script.js"></script>
</body>
</html>
"""
html_file_path = os.path.join(output_dir, "simulation.html")
with open(html_file_path, "w") as html_file:
    html_file.write(html_content)
print(f"HTML structure written to {html_file_path}")
```

## Add CSS for Styling
Define CSS styles for the canvas and other elements to ensure proper layout and appearance.

```python
# Add CSS for Styling
css_content = """
body {
    margin: 0;
    display: flex;
    justify-content: center;
    align-items: center;
    height: 100vh;
    background-color: #f0f0f0;
}

canvas {
    border: 1px solid #000;
}
"""
css_file_path = os.path.join(output_dir, "styles.css")
with open(css_file_path, "w") as css_file:
    css_file.write(css_content)
print(f"CSS styles written to {css_file_path}")
```

## Implement JavaScript for Physics and Animation
Write JavaScript to handle the rotation of the hexagon, simulate gravity, apply friction, and calculate realistic bouncing of the ball.

```python
# Implement JavaScript for Physics and Animation
js_content = """
const canvas = document.getElementById('simulationCanvas');
const ctx = canvas.getContext('2d');

canvas.width = 800;
canvas.height = 600;

let angle = 0;
const hexagonRadius = 200;
const ball = {
    x: canvas.width / 2,
    y: canvas.height / 2,
    radius: 15,
    dx: 2,
    dy: 2,
    gravity: 0.1,
    friction: 0.99
};

function drawHexagon() {
    ctx.save();
    ctx.translate(canvas.width / 2, canvas.height / 2);
    ctx.rotate(angle);
    ctx.beginPath();
    for (let i = 0; i < 6; i++) {
        const x = hexagonRadius * Math.cos((i * Math.PI) / 3);
        const y = hexagonRadius * Math.sin((i * Math.PI) / 3);
        if (i === 0) {
            ctx.moveTo(x, y);
        } else {
            ctx.lineTo(x, y);
        }
    }
    ctx.closePath();
    ctx.stroke();
    ctx.restore();
}

function drawBall() {
    ctx.beginPath();
    ctx.arc(ball.x, ball.y, ball.radius, 0, Math.PI * 2);
    ctx.fillStyle = 'red';
    ctx.fill();
    ctx.closePath();
}

function updateBall() {
    ball.dy += ball.gravity;
    ball.dx *= ball.friction;
    ball.dy *= ball.friction;

    ball.x += ball.dx;
    ball.y += ball.dy;

    if (ball.x + ball.radius > canvas.width || ball.x - ball.radius < 0) {
        ball.dx = -ball.dx;
    }
    if (ball.y + ball.radius > canvas.height || ball.y - ball.radius < 0) {
        ball.dy = -ball.dy;
    }
}

function animate() {
    ctx.clearRect(0, 0, canvas.width, canvas.height);
    drawHexagon();
    drawBall();
    updateBall();
    angle += 0.01;
    requestAnimationFrame(animate);
}

animate();
"""
js_file_path = os.path.join(output_dir, "script.js")
with open(js_file_path, "w") as js_file:
    js_file.write(js_content)
print(f"JavaScript for physics and animation written to {js_file_path}")
```

## Write to HTML File
Combine the HTML, CSS, and JavaScript into a single file and save it to the 'output' folder.

```python
# Write to HTML File
print("All components (HTML, CSS, JS) have been written to the output folder.")
print(f"Open {html_file_path} in a browser to view the simulation.")
```
